```python
# Building the Linear Regression model using ISI, DMC, BUI, DC, FFMC, and Temperature as predictors for FWI,
# reason is: because of its strong correlation with the traget FWI

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Selecting features based on correlation analysis
forestfire_features = ["ISI", "DMC", "BUI", "DC", "FFMC", "Temperature"]
target = "FWI"

# Splitting of the data
x = forestfire[forestfire_features]
y = forestfire[target]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
#model training for linear regression
lr.fit(x_train,y_train)
```

```python
lr.coef_, lr.intercept_
```

```python
y_pred=lr.predict(x_test)
```

```python
w=pd.DataFrame({'Actual':y_test, 'Predicted':y_pred})
```

```python
w['absolute_error']=abs(w['Actual']-w['Predicted'])
w['absolute_error'].mean()
```

```python
w['squared_error']=(w['Actual']-w['Predicted'])**2
```

```python
mean_squared_error(y_test, y_pred)
```

```python
r2_score(y_test, y_pred)
```

```python
# Cross Validation Score
cross_val_scores = cross_val_score(lr, x, y, cv = 5, scoring = 'neg_mean_squared_error')

# Convert scores to positive and get the average RMSE
cross_val_rmse = (-cross_val_scores.mean()) ** 0.5
print(f"Cross-Validated RMSE: {cross_val_rmse}")
```

```python
# Hyper parameter tunning

from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

# Define a range of alpha values
Ridge = Ridge()
Param_grid = {'alpha': [0.01, 0.1, 1, 10, 100]}

# Perform Grid Search with cross-validation
grid_search = GridSearchCV(Ridge, Param_grid, cv = 5, scoring = 'neg_mean_squared_error')
grid_search.fit(x_train, y_train)

# Best parameters and corresponding RMSE
best_alpha_value = grid_search.best_params_['alpha']
best_scores = (-grid_search.best_score_) ** 0.5

print(f"Best alpha: {best_alpha_value}")
print(f"Best Cross-Validated RMSE: {best_scores}")
```
